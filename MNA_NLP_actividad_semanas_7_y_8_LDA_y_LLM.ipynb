# **Ejercicio 1:**

* #### **Liga de los audios de las fábulas de Esopo:** https://www.gutenberg.org/ebooks/21144

* #### **Descargar los 10 archivos de audio solicitados: 1, 4, 5, 6, 14, 22, 24, 25, 26, 27.**



```python
import requests

url = 'https://www.gutenberg.org/files/21144/mp3/21144-' # ##.mp3
filesL = [1, 4, 5, 6, 14, 22, 24, 25, 26, 27]

for f in filesL:
  txSt = f'{f}'
  if f < 10:
    txSt = '0'+txSt
  response = requests.get(url+f'{txSt}.mp3')

  # Check if the request was successful (status code 200)
  if response.status_code == 200:
      # Open a local file in binary write mode ('wb') and save the content
      with open(f'/content/sample_data/21144-{txSt}.mp3', "wb") as file:
          file.write(response.content)
      print("Download complete!")
  else:
      print(f"Failed to download file. Status code: {response.status_code}")
```

# **Ejercicio 2b:**

* #### **Eliminar el inicio y final comunes de los textos extraídos de cada fábula.**

* #### **Sugerencia:** Pueden guardar esta información en un archivo tipo JSON, para que al estar probando diferentes opciones en los ejercicios siguientes, puedan recuperar rápidamente la información de cada video/fábula.

```python
# Incluyan a continuación todas las celdas (de código o texto) que deseen...

import re

nText = []

for t in fText:
  txt = re.sub(r'^.*?f[áa]b[ou]la número \d+(?:,.*?\.)?\.?\s*', '', t,flags=re.IGNORECASE)
  txt = re.sub(r'F[íi]?n\s*de?\s*la\s*f[áa]bula[.,]?\s*.*$', '', txt, flags=re.IGNORECASE | re.DOTALL)

  nText.append(txt.strip())

nText
```

# **Ejercicio 3:**

* #### **Apliquen el proceso de limpieza que consideren adecuado.**

* #### **Justifiquen los pasos de limpieza utilizados. Tomen en cuenta que el texto extraído de cada fábula es relativamente pequeño.**

* #### **En caso de que decidan no aplicar esta etapa de limpieza, deberán justificarlo.**

```python
# Incluyan a continuación todas las celdas (de código o texto) que deseen...

import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stopwords = stopwords.words('spanish')
filesL = [1, 4, 5, 6, 14, 22, 24, 25, 26, 27]
docsTokens = []
docs = []
for f in nText:
  for stw in stopwords:
    f = re.sub(f'\\b{stw}\\b', '', f)
    f = re.sub(r"[.|,]", "", f)
    f = re.sub(r'\s+', ' ', f)
  docs.append(f)
  docsTokens.append([w.lower() for w in f.split(' ')])

for i in docsTokens:
  print(i[:5])
```

```python
# Incluyan a continuación todas las celdas (de código o texto) que deseen...

import gensim
from gensim.corpora import Dictionary
from gensim.models import LdaModel

N = 20  # palabras clave por fábula

for i, fabula in zip(filesL,docsTokens):

    # Diccionario por fábula
    dictionary = Dictionary([fabula])

    # Corpus BOW por fábula
    corpus = [dictionary.doc2bow(fabula)]

    # LDA con 1 tópico
    lda = LdaModel(
        corpus=corpus,
        id2word=dictionary,
        num_topics=1,
        passes=30,
        random_state=42
    )

    # Obtener 20 palabras clave
    topic_words = lda.show_topic(0, topn=N)
    keywords = [word for word, prob in topic_words]

    print(f"Fábula {i} — Top {N} palabras clave:")
    print(keywords)
```

# **Ejercicio 5a y 5b:**

* #### **5a: Mediante el LLM que hayan seleccionado, generar un único enunciado que describa o resuma cada fábula.**

* #### **5b: Mediante el LLM que hayan seleccionado, generar tres posibles enunciados diferentes relacionados con la historia de la fábula.**

* #### **Sugerencia:** En realidad los dos incisos a y b se pueden obtener con un solo prompt que solicite la información y el formato correspondiente para cada una de estas partes. Por ejemplo, para cada fábula la salida puede ser un primer enunciado genérico que resume o describe dicha temática; seguido de tres enunciados, cada uno hablando sobre una situación o parte diferente de la fábula.

```python
# Incluyan a continuación todas las celdas (de código o texto) que deseen...
from transformers import pipeline
from transformers import AutoTokenizer, AutoModelForCausalLM
from huggingface_hub import login


model_id = "meta-llama/Llama-3.2-3B-Instruct"
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    device_map="auto",
)


filesL = [1, 4, 5, 6, 14, 22, 24, 25, 26, 27]
for i, fb in zip(filesL,nText):
  prompt = f"""
Eres un experto en literatura.
Genera lo siguiente de la fábula en el formato:
RESUMEN:
ENUNCIADO 1:
ENUNCIADO 2:
ENUNCIADO 3: [Ultimo elemento]

Fábula:
{fb}
"""



  inputs = tokenizer(prompt, return_tensors="pt", truncation=True).to(model.device)
  outputs = model.generate(**inputs, max_length=300)
  print(f'Fábula {i}:')
  print(tokenizer.decode(outputs[0], skip_special_tokens=True))
  print()
```
