# **Ejercicio 1:**

* #### **Liga de los audios de las fábulas de Esopo:** https://www.gutenberg.org/ebooks/21144

* #### **Descargar los 10 archivos de audio solicitados: 1, 4, 5, 6, 14, 22, 24, 25, 26, 27.**



```python
# Ejercicio 1 — Descargar los 10 audios MP3 de las fábulas de Esopo
import urllib.request
import os

# URLs de los 10 audios requeridos: 1, 4, 5, 6, 14, 22, 24, 25, 26, 27
fabulas = [1, 4, 5, 6, 14, 22, 24, 25, 26, 27]
base_url = "https://www.gutenberg.org/files/21144/mp3/21144-{:02d}.mp3"

os.makedirs('audios', exist_ok=True)

for num in fabulas:
    url = base_url.format(num)
    nombre = f'audios/fabula_{num:02d}.mp3'
    print(f'Descargando fábula {num}...')
    urllib.request.urlretrieve(url, nombre)
    print(f'  Guardado: {nombre}')

print('\nDescarga completa. Archivos disponibles:')
for f in sorted(os.listdir('audios')):
    print(f'  {f}')


```

# **Ejercicio 2a:**

* #### **Comenten el por qué del modelo seleccionado para extracción del texto de los audios.**

* #### **Extraer el contenido de los audios en texto.**

* #### **Sugerencia:** pueden extraerlo en un formato de diccionario, clave:valor $→$ {audio01:fabula01, ...}

Se seleccionó el modelo **openai/whisper-large-v3-turbo** de HuggingFace
porque no requiere API key, es de código abierto, está optimizado para
transcripción en español con múltiples acentos (relevante dado que los
audios provienen de diferentes hablantes hispanos), y maneja audios de
más de 30 segundos automáticamente con `return_timestamps=True`. Es una
versión compacta del modelo whisper-large-v3 con buen balance entre
velocidad y precisión.

```python
!pip install transformers accelerate -q
!apt-get install -y ffmpeg -q

import torch
import warnings
warnings.filterwarnings("ignore")
from transformers import pipeline

# Verificar GPU
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Usando dispositivo: {device}")

# Cargar modelo Whisper large-v3-turbo
model_id = "openai/whisper-large-v3-turbo"
pipe = pipeline(
    "automatic-speech-recognition",
    model=model_id,
    torch_dtype=torch.float32,
    device=device
)

# Transcribir los 10 audios
fabulas = [1, 4, 5, 6, 14, 22, 24, 25, 26, 27]
transcripciones = {}

for num in fabulas:
    audio_path = f'audios/fabula_{num:02d}.mp3'
    print(f'Transcribiendo fábula {num}...')
    result = pipe(
        audio_path,
        return_timestamps=True,
        generate_kwargs={"language": "es"}
    )
    transcripciones[f'audio{num:02d}'] = result["text"]
    print(f'  ✓ Fábula {num} lista')

print('\nTranscripciones completadas.')
print(f'Total de fábulas transcritas: {len(transcripciones)}')


```

# **Ejercicio 2b:**

* #### **Eliminar el inicio y final comunes de los textos extraídos de cada fábula.**

* #### **Sugerencia:** Pueden guardar esta información en un archivo tipo JSON, para que al estar probando diferentes opciones en los ejercicios siguientes, puedan recuperar rápidamente la información de cada video/fábula.

```python
# Ejercicio 2b — Eliminar inicio y final comunes + guardar en JSON

import re
import json

def limpiar_fabula(texto):
    # Eliminar el inicio: "Las fábulas de Esopo/Sopo... fábula número ##"
    # (e)? cubre el caso donde Whisper transcribe "Sopo" en lugar de "Esopo"
    texto = re.sub(
        r'(?i)las\s+f[aá]bulas\s+de\s+(e)?sopo.*?f[aá]bula\s+n[uú]mero\s+\d+\.?\s*',
        '', texto
    )
    # Eliminar el final: "Fin de la fábula, esta grabación..."
    texto = re.sub(
        r'(?i)fin\s+de\s+la\s+f[aá]bula.*$',
        '', texto
    )
    return texto.strip()

# Aplicar limpieza a todas las transcripciones
fabulas_limpias = {}
for clave, texto in transcripciones.items():
    fabulas_limpias[clave] = limpiar_fabula(texto)
    print(f'{clave}:')
    print(f'  {fabulas_limpias[clave][:100]}...')
    print()

# Guardar en JSON para no repetir la transcripción
with open('fabulas.json', 'w', encoding='utf-8') as f:
    json.dump(fabulas_limpias, f, ensure_ascii=False, indent=2)

print('Archivo fabulas.json guardado correctamente.')
print(f'Total de fábulas: {len(fabulas_limpias)}')
```

# **Ejercicio 3:**

* #### **Apliquen el proceso de limpieza que consideren adecuado.**

* #### **Justifiquen los pasos de limpieza utilizados. Tomen en cuenta que el texto extraído de cada fábula es relativamente pequeño.**

* #### **En caso de que decidan no aplicar esta etapa de limpieza, deberán justificarlo.**

Dado que cada fábula tiene aproximadamente 150 palabras después de
eliminar el inicio y final comunes, se aplica una limpieza mínima:

1. **Minúsculas** — para unificar palabras iguales con diferente capitalización.
2. **Eliminar puntuación y caracteres especiales** — no aportan información
   semántica al LDA.
3. **Eliminar stopwords en español** — palabras muy frecuentes que no
   distinguen temas (el, la, de, que, y...).
4. **Tokens de longitud > 2** — elimina residuos de una o dos letras.

Se decidió **NO aplicar stemming ni lematización** porque al ser textos
tan cortos, perder la forma original de las palabras reduciría la
interpretabilidad de las palabras clave extraídas por LDA.

```python
# Ejercicio 3 — Limpieza mínima de los textos

import re
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')
stop_words = set(stopwords.words('spanish'))

def limpiar_texto(texto):
    # 1. Minúsculas
    texto = texto.lower()
    # 2. Solo caracteres alfabéticos
    texto = re.sub(r'[^a-záéíóúüñ\s]', '', texto)
    # 3. Tokenizar
    tokens = texto.split()
    # 4. Eliminar stopwords y tokens cortos
    tokens = [t for t in tokens if t not in stop_words and len(t) > 2]
    return tokens

# Cargar desde JSON por si se reinició la sesión
import json
with open('fabulas.json', 'r', encoding='utf-8') as f:
    fabulas_limpias = json.load(f)

# Aplicar limpieza
fabulas_tokens = {}
for clave, texto in fabulas_limpias.items():
    fabulas_tokens[clave] = limpiar_texto(texto)
    print(f'{clave}: {len(fabulas_tokens[clave])} tokens')
    print(f'  {fabulas_tokens[clave][:10]}')
    print()

print('Limpieza completada.')

```

# **Ejercicio 4:**

```python
# Ejercicio 4 — Extracción de palabras clave con LDA

!pip install gensim -q

from gensim import corpora, models

# Crear el diccionario y corpus para LDA
diccionario = corpora.Dictionary(fabulas_tokens.values())
corpus = [diccionario.doc2bow(tokens) for tokens in fabulas_tokens.values()]

# Parámetros LDA
NUM_TOPICS = 1      # una fábula = un tópico
NUM_PALABRAS = 20   # palabras clave por tópico

# Entrenar un modelo LDA por cada fábula
fabulas_keywords = {}
claves = list(fabulas_tokens.keys())

for i, clave in enumerate(claves):
    tokens = fabulas_tokens[clave]

    # Diccionario y corpus individual por fábula
    dic_fab = corpora.Dictionary([tokens])
    corpus_fab = [dic_fab.doc2bow(tokens)]

    # Modelo LDA
    lda = models.LdaModel(
        corpus=corpus_fab,
        id2word=dic_fab,
        num_topics=NUM_TOPICS,
        random_state=42,
        passes=50,
        iterations=400
    )

    # Extraer palabras clave del único tópico
    palabras = lda.show_topic(0, topn=NUM_PALABRAS)
    keywords = [palabra for palabra, prob in palabras]
    fabulas_keywords[clave] = keywords

    print(f'{clave}:')
    print(f'  {keywords}')
    print()

print('Palabras clave extraídas para las 10 fábulas.')
```

# **Ejercicio 5a y 5b:**

* #### **5a: Mediante el LLM que hayan seleccionado, generar un único enunciado que describa o resuma cada fábula.**

* #### **5b: Mediante el LLM que hayan seleccionado, generar tres posibles enunciados diferentes relacionados con la historia de la fábula.**

* #### **Sugerencia:** En realidad los dos incisos a y b se pueden obtener con un solo prompt que solicite la información y el formato correspondiente para cada una de estas partes. Por ejemplo, para cada fábula la salida puede ser un primer enunciado genérico que resume o describe dicha temática; seguido de tres enunciados, cada uno hablando sobre una situación o parte diferente de la fábula.

Se seleccionó la API de **Claude (claude-sonnet-4-20250514)** como modelo LLM
porque genera respuestas estructuradas de alta calidad en español, permite
controlar el formato de salida con precisión, y en un solo prompt produce
tanto el resumen general como los tres subtemas de cada fábula.

```python
!pip install groq -q

import time
from groq import Groq
from google.colab import userdata

client_groq = Groq(api_key=userdata.get('GROQ_API_KEY'))

def generar_resumen_subtemas(keywords):
    prompt = f"""Eres un experto en literatura clásica.
Con base en las siguientes palabras clave de una fábula de Esopo:
{', '.join(keywords)}

Genera exactamente lo siguiente en español:
RESUMEN: Un único enunciado que describa o resuma la fábula.
SUBTEMA 1: Un enunciado sobre un aspecto de la fábula.
SUBTEMA 2: Un enunciado sobre otro aspecto diferente.
SUBTEMA 3: Un enunciado sobre un tercer aspecto diferente.

Responde SOLO con ese formato, sin explicaciones adicionales."""

    response = client_groq.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=[{"role": "user", "content": prompt}],
        max_tokens=300
    )
    return response.choices[0].message.content

# Generar para las 10 fábulas
fabulas_llm = {}

for clave, keywords in fabulas_keywords.items():
    print(f'Procesando {clave}...')
    resultado = generar_resumen_subtemas(keywords)
    fabulas_llm[clave] = resultado
    print(f'\n{clave}:')
    print(resultado)
    print('-'*50)
    time.sleep(3)

print('\nGeneración completada para las 10 fábulas.')
```

### Conclusiones de la Actividad Audio-a-Texto: LDA y LLM

**Pipeline implementado:**
La actividad integró tres tecnologías: transcripción automática de audio
(Whisper), modelado de tópicos (LDA) y generación de lenguaje (LLaMA 3.1 via Groq),
formando un pipeline completo de análisis de contenido de audio.

**Principales hallazgos:**

1. **Transcripción con Whisper** — El modelo whisper-large-v3-turbo demostró
buena precisión en español con múltiples acentos. El mayor reto fue el audio 25,
correspondiente al hablante no nativo de español, donde la transcripción presentó
mayor número de errores léxicos.

2. **Limpieza de texto** — Se confirmó la advertencia del PDF: al ser textos muy
cortos (~150 palabras por fábula), una limpieza agresiva elimina información
valiosa. La decisión de aplicar limpieza mínima fue acertada para preservar el
vocabulario necesario para el LDA.

3. **LDA con textos cortos** — El mayor problema técnico fue la longitud reducida
de cada fábula. LDA funciona mejor con documentos largos y vocabulario amplio.
Con ~80-100 tokens por fábula después de la limpieza, el modelo requirió muchas
iteraciones (passes=50) para converger y algunas palabras clave resultaron poco
representativas del tema central.

4. **LLM como complemento** — El uso de LLaMA 3.1 para generar resúmenes y subtemas
a partir de las palabras clave del LDA demostró ser muy efectivo. El LLM compensó
las limitaciones del LDA en textos cortos, produciendo enunciados coherentes y
contextualmente relevantes incluso cuando las palabras clave no eran perfectas.

**Problema principal enfrentado:**
La combinación de textos cortos + LDA es el mayor desafío de la actividad. LDA
asume que los documentos contienen múltiples tópicos mezclados, pero forzarlo a
un solo tópico en textos muy cortos limita su capacidad de discriminación. Una
alternativa sería usar TF-IDF directamente para extraer palabras clave en lugar
de LDA cuando los documentos son tan pequeños.

**Conclusión general:**
La combinación de modelos especializados (Whisper para audio, LDA para tópicos,
LLM para generación) permite construir pipelines robustos de análisis de contenido
sin necesidad de datos etiquetados. Cada modelo cubre las limitaciones del anterior,
siendo el LLM el que mayor valor agregado aporta al final del pipeline.
