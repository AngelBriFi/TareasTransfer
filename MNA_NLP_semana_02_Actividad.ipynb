*   **Pregunta 1.**

Busca y elimina todos los saltos de línea '\n' que se encuentran al final de cada comentario.

Una vez finalizado, imprime los primeros 10 comentarios del resultado obtenido.


```python
#Eliminamos saltos con re.sub
docsSinSaltos = [re.sub(r'\n+$', '', comentario) for comentario in docs]
```

```python
#Resultados obtenidos
docsSinSaltos[0:10]
```

*   **Pregunta 2.**  

Busca e imprime todas las palabras que terminan con dos o más signos de admiración seguidos, por ejemplo "!!!".

Debes imprimir tanto la palabra como la totalidad de signos de admiración que le siguen.

Indica cuántos resultados obtuviste.



```python
multiplesSignos = 0
patron = r'(\b\w+)(!{2,})' #El patron contiene dos regex, la primera obtiene palabras y el segundo verifica si la palabra tiene mas de 2 signos de exclamación
for frase in docsSinSaltos:
  for palabra, signos in re.findall(patron, frase): #Se obtiene la palabra y los signos en una lista
    multiplesSignos += 1
    print(palabra,'---',signos,'---',len(signos))
```

```python
print(f"Se encontraron: {multiplesSignos} casos con multiples signos de amiración")
```

*   **Pregunta 3.**  

Busca e imprime todas las palabras que están escritas totalmente en mayúsculas. Cada coincidencia debe ser una sola palabra.

Indica cuántas palabras encontraste.



```python
todoMayus = 0
patron = r"\b[A-Z]{2,}+\b" #Con 2 o mas letras para evitar frases que contengan I o A, al eliminar {2,} podemos obtener dichos resultados
for frase in docsSinSaltos:
  for palabra in re.findall(patron, frase):
    todoMayus += 1
    print(palabra)
```

```python
print(f"Se encontraron: {todoMayus} casos de palabras totalmente en mayusculas")
```

*   **Pregunta 4.**  

Busca e imprime los comentarios en donde todos los caracteres alfabéticos (letras) están en mayúsculas.

Cada coincidencia encontrada debe ser todo el comentario/enunciado.

Indica cuántos resultados obtuviste.


```python
fraseConPalMayus = 0
patron = r"\b[A-Z]{2,}+\b" #Mismo patro que en Pregunta 3.
for frase in docsSinSaltos:
  if len(re.findall(patron, frase)) > 0: #Ahora solo se verifica que la frase contenga dicho patron para imprimir frase completa
    fraseConPalMayus += 1
    print(frase)
```

```python
print(f"Se encontraron: {fraseConPalMayus} casos de frases con palabras totalmente en mayusculas")
```

*   **Pregunta 5.**  

Busca e imprime todas las palabras que tengan una vocal acentuada, del tipo á, é, í, ó, ú.

Indica cuántos resultados obtuviste.

```python
palabraAcent = 0
aPal=[]
patron = r'\b\w*[áéíóúÁÉÍÓÚ]\w*\b' #Vocales acentuadas en minuscula y mayuscula, tambien se puede utilizar re.ignorecase dentro de findall
for frase in docsSinSaltos:
  for palabra in re.findall(patron, frase):
    aPal.append(palabra)
    palabraAcent += 1
    print(palabra)
```

```python
print(f"Se encontraron: {palabraAcent} casos de palabras que tengan una vocal acentuada, del tipo á, é, í, ó, ú")
```

*   **Pregunta 6.**  

Busca e imprime todas las cantidades numéricas monetarias, enteras o con decimales, que inician con el símbolo $\$$.

Indica cuántos resultados obtuviste.

```python
cantidadesNum = 0
patron = r"\$\d+(?:\.\d+)?" #El simbolo $ necesita ser escapado \$ para indicar que buscamos el simbolo en el texto, seguido de \d+ donde se expresa 1 o mas digitos, para obtener la parte decimal se utiliza ?:\. donde se indica que de encontrar coincidencia de la primera parte mantenga lo siguiente a la derecha del punto o \d+
for frase in docsSinSaltos:
  for palabra in re.findall(patron, frase):
    cantidadesNum += 1
    print(palabra)
```

```python
print(f"Se encontraron: {cantidadesNum} casos de cantidades numéricas monetarias")
```

*   **Pregunta 7.**  

Busca e imprime todas las palabras que sean variantes de la palabra "love", sin importar si incluyen mayúsculas o minúsculas, o la manera en que esté conjugada o alguna otra variación que se haga con dicha palabra.

Indica cuántos resultados obtuviste.

```python
palabrasLov = 0
patron = r"\b(lov\w*)\b" #Se busca la base de la palabra lov, indicando que sea palabra completa \b...\b con cualquier cantidad de caracteres \w*
for frase in docsSinSaltos:
  for palabra in re.findall(patron, frase, re.IGNORECASE):
    palabrasLov += 1
    print(palabra)
print(f"Se encontraron: {palabrasLov} veces love y/o conjugaciones")
```

*   **Pregunta 8.**  

Busca e imprime todas las palabras, variantes de "so" y "good", que tengan dos o más "o" en "so" y 3 o más "o" en good.

Indica cuántas encontraste.


```python
dosMasO = 0
patron = r"\b(so{2,}|go{3,}d{2,})\b" #Se aplica una busqueda con operador or | el primer caso si la palabra contine 2 o mas o despues de so, y la segunda parte si despues de go hay 3 o mas o y si la palabra tiene 1 o mas d (se encontro gooodd)
for frase in docsSinSaltos:
  for palabra in re.findall(patron, frase, re.IGNORECASE):
    dosMasO += 1
    print(palabra)
print(f"Se encontraron: {dosMasO} variantes de \"so\" y \"good\", que tengan dos o más \"o\" en \"so\" y 3 o más \"o\" en good")
```

*   **Pregunta 9.**  

Busca e imprime todas las palabras que tengan una longitud mayor estrictamente a 10 caracteres alfabéticos.

No se consideran los signos de puntuación o caracteres especiales en la longitud de estas cadenas, solo caracteres alfabéticos en mayúsculas o minúsculas.

Indica la cantidad de palabras encontradas.


```python
diezOMas = 0
patron = r"\b[a-zA-Z]{11,}\b" #En la palabra se busca que tenga 11 o mas caracteres
for frase in docsSinSaltos:
  for palabra in re.findall(patron, frase):
    diezOMas += 1
    print(palabra)
print(f"Se encontraron: {diezOMas} palabras que con longitud mayor a 10 caracteres alfabéticos")
```

*   **Pregunta 10.**  

Busca e imprime todas las palabras que inician con una letra mayúscula y terminan con una minúscula, pero que además no sea la primera palabra del comentario/string.

Indica la cantidad de resultados obtenidos.

```python
mayusMin = 0
patron = r'\b[A-Z][a-zA-Z]*[a-z]\b'#Se busca que el inicio de la palabra sea mayuscula seguida de letras minusculas o mayusculas, y se asegura que la ultima palabra sea minuscula
for frase in docsSinSaltos:
  frase=frase.replace('"','') # Eliminar comillas de palabras
  fraseSpl = frase.split() #Creacion de lista de la frase
  fraseSinPrimera = ' '.join(fraseSpl[1:]) #Frase unida omitiendo primera palabra
  for palabra in re.findall(patron, fraseSinPrimera):
    mayusMin += 1
    print(palabra)
print(f"Se encontraron: {mayusMin} palabras que inician con una letra mayúscula y terminan con una minúscula, pero que además no sea la primera palabra del comentario/string")
```

*   **Pregunta 11.**  

Busca e imprime la secuencia de dos o más palabras que están separadas por un guion, "-", sin que tengan espacios en blanco entre ellas.

Por ejemplo "Go-Kart" sería válido, pero "Go  -Kart" o "Go  -  Kart" no lo serían.

Indica la cantidad de resultados obtenidos.

```python
palabrasGuion = 0
patron = r"\b[a-zA-Z]+(?:-[a-zA-Z]+)+\b" #La palabra puede iniciar con minuscula o mayuscula y puede ser cualquier cantidad de caracteres, despues si hay coincidencia (?:-[a-zA-Z]+)+ se encarga de añadir la segunda verificacion que contenga el guion segudido de caracteres, esto elimina el riesgo del espacio entre el guion
for frase in docsSinSaltos:
  for palabra in re.findall(patron, frase):
    palabrasGuion += 1
    print(palabra)
print(f"Se encontraron: {palabrasGuion} palabras que están separadas por un guion, \"-\", sin que tengan espacios en blanco entre ellas")
```

*   **Pregunta 12.**  

Busca e imprime todas las palabras que terminan en "ing" o "ed".

Indica la cantidad de palabras que encontraste de cada una.

```python
finIngEd = 0
patron = r"\b[a-zA-Z]+(?:ing|ed)"
for frase in docsSinSaltos:
  for palabra in re.findall(patron, frase):
    finIngEd += 1
    print(palabra)
print(f"Se encontraron: {finIngEd} palabras que terminan en \"ing\" o \"ed\"")
```

#**Parte 3. Proceso de limpieza.**

*   **Pregunta 13.**  

Ahora realiza un proceso de limpieza del corpus que incluya los siguientes procesos:

*   Solo se deben considerar caracteres alfabéticos. Es decir, se eliminan todos los signos de puntuación y caracteres especiales.
*   Todos los caracteres alfabéticos se transforman a minúsculas.
*   Se deben eliminar todos los espacios en blanco adicionales que se puedan encontrar en cada comentario.

Al finalizar dicho proceso de limpieza, imprime el resultado de los primeros 10 comentarios resultantes.
   




```python
def limpiar_texto(texto):
    texto = texto.lower() #Texto en minusculas
    texto = re.sub(r'[^a-z]', ' ', texto) #Eliminacion de caracteres especiales
    texto = re.sub(r'\s+', ' ', texto) #Eliminacion de espacios extra
    texto = texto.strip() # Eliminacion de expacios en blanco al inicio y fin de la frase

    return texto
```

```python
corpus_limpio = [limpiar_texto(comentario) for comentario in docs]

for i, comentario in enumerate(corpus_limpio[:10], start=1):
    print(f"{i}. {comentario}")
```

*   **Pregunta 14.**  

Con el resultado de la limpieza obtenido en la pregunta anterior, realiza ahora un proceso de tokenización por palabras del corpus.

Es decir, al final de este proceso de tokenización, debes tener como resultado una lista de listas, donde cada comentario estará tokenizado por palabras.

Al terminar calcula el total de tokens obtenido en todo el corpus.

```python
# Tokenizar el corpus
corpus_tokenizado = [frase.split() for frase in corpus_limpio]

# Total de tokens obtenidos
total_tokens = sum(len(frase) for frase in corpus_tokenizado)

# Ejemplos de resultados obtenidos
for i, frase in enumerate(corpus_tokenizado[:5], start=1):
    print(f"{i}. {frase}")

print("\nTotal de tokens en el corpus:", total_tokens)
```

*   **Pregunta 15.**  

Finalmente, en este ejercicio definiremos nuestro conjunto de palabras "stopwords", las cuales deberás eliminar de todo el corpus.

Recuerda que ejemplos de stopwords son artículos, adverbios, conectivos, etcétera, que tienen frecuencias de aparición muy altas en cualquier documento, pero que no brindan mucho significado en cuanto al significado de un enunciado.

Con base a la lista de stopwords que se te proporciona, realiza un proceso de limpieza eliminando todas estas palabras del corpus obtenido en el ejercicio anterior.

Obtener cuántos tokens/palabras quedan finalmente en todo el corpus.

Obtener cuántos de estos tokens/palabras son diferentes, es decir, cuántos tokens únicos tendrá lo que llamaremos más adelante nuestro vocabulario.

```python
# Considera la siguiente lista como tu conjunto de stopwords:
mis_stopwords = ['i', 'me', 'my', 'myself', 'we', 'our', 'ours', 'ourselves', 'you', 'your', 'yours', 'he', 'him', 'his', 'himself', 'she', 'her', 'hers', 'herself', 'it', 'its', 'itself', 'they', 'them', 'their', 'theirs', 'themselves', 'what', 'which', 'who', 'whom', 'this', 'that', 'these', 'those', 'am', 'is', 'are', 'was', 'were', 'be', 'been', 'being', 'have', 'has', 'had', 'having', 'do', 'does', 'did', 'doing', 'a', 'an', 'the', 'and', 'but', 'if', 'or', 'because', 'as', 'until', 'while', 'of', 'at', 'by', 'for', 'with', 'about', 'against', 'between', 'into', 'through', 'during', 'before', 'after', 'above', 'below', 'to', 'from', 'up', 'down', 'in', 'out', 'on', 'over', 'under', 'again', 'further', 'then', 'once', 'here', 'there', 'when', 'where', 'why', 'how', 'all', 'any', 'both', 'each', 'few', 'more', 'most', 'other', 'some', 'such', 'only', 'own', 'same', 'so', 'than', 'too', 'very', 's', 't', 'can', 'will', 'just', 'should', 'now', 'll']
```

```python
def eliminar_stopwords(tokens):
    return [token for token in tokens if token not in mis_stopwords]
```

```python
# Aplicar eliminación de stopwords a todo el corpus
corpus_sin_stopwords = [eliminar_stopwords(comentario) for comentario in corpus_tokenizado]

# Calcular total de tokens finales
total_tokens_final = sum(len(comentario) for comentario in corpus_sin_stopwords)

# Calcular vocabulario único
vocabulario = set([token for frase in corpus_sin_stopwords for token in frase])
total_vocabulario = len(vocabulario)

# Mostrar primeros 10 comentarios sin stopwords
for i, comentario in enumerate(corpus_sin_stopwords[:10], start=1):
    print(f"{i}. {comentario}")

print("\nTotal de tokens después de eliminar stopwords:", total_tokens_final)
print("Total de tokens únicos (vocabulario):", total_vocabulario)
```
